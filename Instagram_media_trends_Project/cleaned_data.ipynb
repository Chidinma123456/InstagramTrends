```python
df = df.drop(['Impressions', 'Campaign ID', 'Sentiment', 'Influencer ID', 'Audience Gender', 'Post ID'], axis=1)
```

```python
df = df.drop(['Audience Interests'], axis =1)
```

```python
df = df[df['Platform'] == 'Instagram']
```

```python
df
```

```python
# make sure the timestamp column is in datetime format
df.loc[:, 'Post Timestamp'] = pd.to_datetime(df['Post Timestamp'])
```

```python
# Create separate columns for date and time
df.loc[:, 'date'] = df['Post Timestamp'].dt.date  # Extract the date part
df.loc[:, 'time'] = df['Post Timestamp'].dt.time  # Extract the time part
```
